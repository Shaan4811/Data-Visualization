## deaths over cases in first 20 countries

```python
country.head(1)
```

```python
first20=country.sum(numeric_only=True).head(20)
```

```python
first20
```

```python
plt.plot(first20['cases'],first20['deaths'],marker='o')
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
plt.bar(first20['cases'],first20['deaths'],color='r',edgecolor='black')
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
a=first20[['cases','deaths']]
```

```python
a
```

```python
a.plot(kind='line',colormap='plasma',figsize=(8,6),marker='o')
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
a.plot(kind='bar',colormap='plasma',edgecolor='black',stacked=True)
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('deaths over cases');
```

```python
a.plot(kind='hist',colormap='plasma');
```

```python
a.hist(bins=10,edgecolor='black',color='r')
plt.suptitle('cases and deaths rate')
plt.grid(False);
```

```python
plt.boxplot([first20['cases'],first20['deaths']]);
```

```python
a.boxplot(labels=a.columns);
```

```python
plt.violinplot([first20['cases'],first20['deaths']])
plt.xlabel('cases')
plt.ylabel('deaths')
plt.title('cases and deaths rate');
```
