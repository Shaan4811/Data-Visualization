## Description

```python
# Unnamed: 0: Likely an index column or a redundant ID column.
# id: The unique identifier for each movie or data entry.
# budget: The allocated budget for the movie, likely in millions.
# popularity: A measure of how popular the movie is, based on factors like views, mentions, etc.
# revenue: The total earnings or revenue the movie generated, possibly in millions.
# title: The title or name of the movie.
# vote_average: The average rating the movie received, usually out of 10 (scaled to 100 here).
# vote_count: The number of people who voted or rated the movie.
# director_id: The unique identifier for the movie's director.
# year: The year the movie was released or produced.
# month: The month the movie was released or produced.
# day: The day of the month the movie was released or produced.
```

## q1: retrieve which movie has the highest revenue and list top 10 highest revenue genrated movies

```python
movdir['revenue'].max()
```

```python
revenue=movdir.sort_values(by='revenue',ascending=False).reset_index(drop=True).head(10)
```

```python
revenue
```

```python
top10revenue=revenue[['title','revenue']]
```

```python
top10revenue
```

```python
plt.figure(figsize=(10,5))
plt.plot(top10revenue['title'],top10revenue['revenue'],marker='o',c='r')
plt.xticks(rotation=90)
plt.xlabel('Name')
plt.ylabel('Revenues')
plt.grid(True)
plt.title('Top 10 revenue generating movies');
```
